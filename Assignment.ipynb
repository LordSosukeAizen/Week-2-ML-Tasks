# Random Forests and Decision Trees
- Classify the Spotify.csv dataset 
- Predict the Genre 
- The dataset has 23 Columns and The Output feature is **playlist_genre** so all others can be used as input features. Decide on what can be used and what cannot be used. 
- Do the Preprocessing below
- Do 70:30 Train test split 

```python
import pandas as pd 
main_data = pd.read_csv('spotify.csv')
view_data = pd.read_csv('spotify Edited.csv')
data = pd.DataFrame(main_data.values, columns=view_data.columns)
```

```python
data.head()
```

```python
data.columns
```

```python
# data['playlist_subgenre'].unique()
```

```python
unwanted_columns = ['track_id', 'track_name', 'track_popularity', 'track_album_id', 'track_album_name', 
                    'track_album_release_date','playlist_id', 'track_artist']
```

```python
data.drop(columns=unwanted_columns, inplace=True)
```

```python
data.head()
```

```python
# No naN values is the data.
print(data.isnull().sum()) ; print('\n' ,len(data))
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler

Lbl = LabelEncoder()
one_hot_encoder = OneHotEncoder()
# std = StandardScaler()

data['playlist_subgenre'] = Lbl.fit_transform(data['playlist_subgenre'])
data['playlist_name'] = Lbl.fit_transform(data['playlist_name'])
data['playlist_genre'] = Lbl.fit_transform(data['playlist_genre'])
# y = one_hot_encoder.fit_transform(data['playlist_genre'])
y = data['playlist_genre']
X = data.drop(columns=['playlist_genre'], axis=1)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Frame a Decision Tree Model 
- Use Any Parameters
- You can use any library

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
predicted = clf.predict(X_test)
```

```python
Lbl.inverse_transform(predicted)
```

```python
Lbl.inverse_transform(y_test)
```

## Frame a Random Forests Model 
- Use Any Parameters
- You can use any library

```python
from sklearn.ensemble import RandomForestClassifier

clf2 = RandomForestClassifier()
clf2.fit(X_train, y_train)
clf2_predicted = clf2.predict(X_test)
```

```python
Lbl.inverse_transform(clf2_predicted)
```

```python
Lbl.inverse_transform(y_test)
```

## Train the Models and Report the Accuracy

```python
from sklearn.metrics import accuracy_score

print(f'accuracy score for Decision Tree classifier : {accuracy_score(y_test, predicted)}')
print(f'accuracy score for Random Forest classifier : {accuracy_score(y_test, clf2_predicted)}')
```
